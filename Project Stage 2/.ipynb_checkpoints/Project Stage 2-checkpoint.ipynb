```python
# Groups specific columns of the data by the weighted_price 
# Finds the sum of the BTC volumn 
# This allows us to know where BTC trades occurred the most.

columns = dataframe.loc[:,['vol_aud', 'weighted_price']]
columns = (columns / 100)
columns = columns.round(0) * 100
btc_occurrences = columns.groupby(['weighted_price']).sum()

# Point of control: aka most voluminous areas of the BTC prices
fig, ax = plt.subplots()
ax.plot(btc_occurrences)
ax.set_facecolor('white')
ax.ticklabel_format(useOffset=False, style='plain')
plt.ylabel('Volume in AUD')
plt.xlabel('Price')
plt.title('Total AUD volume at different prices between 2014-2016')
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
btc_occurrences.plot(style='b', ax=ax, kind = 'bar')
pass
```

```python
# LTCBTC graph
fig = plt.figure(figsize=(18,10))
ltc_graphed = dataframe.set_index('Date')
ltc_graphed.plot(y = 'LTCBTC', kind = 'area', color = 'purple')
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
pass
plt.xlabel('Date', fontsize=15, fontname='Times New Roman')
plt.ylabel('Price', fontsize=15, fontname='Times New Roman')
plt.title('The price of LTC vs BTC between 2014-2016', fontsize=20, fontname='Times New Roman')
```

```python
candles_btc = dataframe.drop(['LTCBTC','vol_aud','vol_btc', 'weighted_price'], 1)[0:591]

candles_btc.columns = ['Date', 'Open', 'High', 'Low', 'Close']
candles_btc['Date'] = candles_btc['Date'].map(lambda x: pd.to_datetime(x, dayfirst=True)).map(mdates.date2num)

fig = plt.figure(figsize=(18,10))
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=6, colspan=1)

#Converts raw mdate numbers to dates
ax1.xaxis_date()
ax1.set_facecolor('gray')
#Making candlestick plot
candlestick_ohlc(ax1,candles_btc.values,width=3, colorup='g', colordown='r',alpha=0.75)
#plt.legend('BTCAUD')
plt.xlabel('Date', fontsize=20, fontname='Times New Roman')
plt.ylabel('Price', fontsize=20, fontname='Times New Roman')
plt.title('BTC vs AUD between 2014-2016', fontsize=25, fontname='Times New Roman')
#plt.text(1.1,12, r'$y = x^2$', fontsize=15, bbox={'facecolor':'yellow'})
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
pass
```
